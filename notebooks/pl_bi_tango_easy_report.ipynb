### Helping PMI Organize Their Emergency Warehouse
*By: Tango Team (Jeffri,Wira,Natassa)*

<div style="text-align: justify"> Based on our assessment, “goods storing practice happened when the goods are easily found and neatly categorized in one place” which also in line with the "10 Principals for best warehouse design and operation" popularized by Mal Walker especially on the third point; “Stock is in the right place”
This thought then bring us to the PMI Emergency Warehouse Project that expected to help PMI organized their warehouse. </div>


#### Problem Analysis
Before jumping to the next step, our team tries to understand the current condition. The facts are
-	The warehouse has 7 segments in Jakarta (unfortunately there are no other specific information of the segmentation)
-	The warehouse is functioned as Covid-19 Emergency Warehouse which has various type from medical equipment, medicine, disinfectants, F&B, etc.
-	Turn overtime is high. There’s one case where the aging of the goods only stays at 5 hours!


<div style="text-align: justify"> The issue itself triggered as we found some of the goods aren’t in the right place and mixed up on another warehouses, take examples, 1) food segment is scattered in three different warehouses 2) some of the disinfectants are in the same warehouse as food and beverage which might have violating WHO Good Storage Practice and PMI’s logistics manual that stated “Do not store fuel, chemicals, fertilizers, pesticides or cement in the same place as food” (PMI’s logistics manual point 2.4). On the other side, we are aware this condition is likely to happen with such high turnaround time. 
 </div>

#### User Needs
Assumed that we, Tango Team, hired as the warehouse manager in PMI’s emergency warehouse, things most likely to be asked are:
1.	I got order to send goods to the impacted area. Where can I find the particular goods? Let’s say, foods, medical equipment, evacuation equipment, or etc?
2.	I need a certain amount of goods. How many are these goods available in the warehouse? Does it enough to cover the orders? Are they concentrated at one warehouse?
3.	Do the warehouses have the categorization?
4.	Can I easily search where the warehouse that store Face Mask? Or maybe Books? Or, can I easily see what’s inside warehouse 1?


#### Ideation
From the queries we formulated, a dashboard might help PMI organize item placement in effective and efficient way. Therefore, we make a dashboard that can centralized PMI Emergency Warehouse Data called **WeCat** (Warehouse Categorization). This solution expected to help warehouse manager to give reallocation suggestion of misplaced item and give sense of prioritization in organizing item placement across emergency warehouse.

#### The Features and How to Use It
1.	The Stock Accuracy as the success metrics indicator &#8594; the monitoring indicator to know the items that already placed in the right warehouse


2.	Item Search and date filter &#8594; give user easiness to navigate through the data. User can directly search the Nama Barang, Group Barang, and Gudang Awal to make it easier to reallocate items in item search and move the date slider to search a certain date range.


3.	Reallocation by Warehouse Matrix &#8594; gives user the sense of priority to the goods which needed to be reallocate based on weight balance (balance in – balance out). Ideally user should reallocate the top-10 items first before jumping on the next item.

 
4.	Gudang Standard Mapping &#8594; gives a guideline for user to know the categorization of each warehouse. Can be customize according to PMI’s needs


5.	Group Barang Item Monitor &#8594; gives a guideline of each Group Barang items. Easily filtered using dropdown filter
